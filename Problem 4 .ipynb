```python
# Here we are importing all the libraries
#from pymnet import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import networkx as nx
```

```python
data = pd.read_csv("Assignment_1_Data.csv")
data.head(2)
```

```python
#get the name of students
name_of_students = data.iloc[:,1]
name_of_students_list = name_of_students.values.tolist()
#get the course name
clean_data = data.iloc[:,2:]
name_of_the_courses = list(clean_data)
name_of_the_courses
# make a list of last the columns 
clean_data_list = clean_data.values.tolist()
clean_data_list
```

```python


def draw_graph(graph):

    # extract nodes from graph
    nodes = set([n1 for n1, n2 in graph] + [n2 for n1, n2 in graph])

    # create networkx graph
    G=nx.Graph()

    # add nodes
    for node in nodes:
        G.add_node(node)

    # add edges
    for edge in graph:
        G.add_edge(edge[0], edge[1])

    # draw graph
    pos = nx.shell_layout(G)
    nx.draw(G, pos)

    # show graph
    plt.show()
```

```python
#Get row and column
row , col = clean_data.shape
```

```python
name_of_students = [*range(0, row, 1)]

for k in range(0, col):
    list_for_touple = []
    for i in range(0, row):
        for j in range(i,row):
            if clean_data_list[i][k] == clean_data_list[j][k] and i !=j:
                connected_data = (name_of_students[i], name_of_students[j])
                list_for_touple.append(connected_data)
            
    draw_graph(list_for_touple)
    print(list_for_touple)
    print('\n')
    com_glo_parameter(list_for_touple)
   
    #         else:
    #             print("NO")
```

```python
def com_glo_parameter(list_for_touple):
    G = nx.Graph(list_for_touple)
    if nx.is_connected(G) == True:
        G = nx.Graph(list_for_touple)

        #size of the graph
        print(G.size(weight="weight"))

        #diameter of the graph
        print(nx.diameter(G))

        #average  path length
        print(nx.average_shortest_path_length(G))

        #average clustering coefficient
        print(nx.average_clustering(G))
        
    else:
        largest_cc = max(nx.connected_components(G), key=len) 

        diconnected_graph(list_for_touple, largest_cc)
        
     
        
        
```

```python
def diconnected_graph(list_for_touple, largest_cc):    
    list_of_touple_largest_graph = []
    for i in range(0, len(list_for_touple)):
        if list_for_touple[i][0] in largest_cc:
            list_of_touple_largest_graph.append(list_for_touple[i])
    print(list_of_touple_largest_graph)
    G = nx.Graph(list_of_touple_largest_graph)
    
    #average shortest path
    print(nx.average_shortest_path_length(G))
    
    #size of the graph
    print(G.size())

    #diameter of the graph
    print(nx.diameter(G))
    
    #average clustering coefficient
    print(nx.average_clustering(G))
```
