```python
# Here we are importing all the libraries
from pymnet import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Problem 1

```python
#Upload the csv file via pandas
assignment_1_data = pd.read_csv("Assignment_1_Data.csv")
assignment_1_data.head(2)
```

```python
#Get an overview of the data
assignment_1_data.describe()
```

```python
#lets drop the unnecessary colums , keep the course column
assignment_1_data_clean = assignment_1_data.drop(['Timestamp', 'Student name'], axis=1)
assignment_1_data_clean.head(2)
```

```python
#column header name , we may need to use them in future
column_headers = list(assignment_1_data_clean)
column_headers
```

```python
#Convert the values into dummy values so it will be easier
dummies = pd.get_dummies(assignment_1_data_clean)

dummies_clean = dummies.iloc[: , 1:]
dummies_clean_old =dummies_clean
dummies_clean
```

```python
#convert the pandas data frame into numpy, numpy is easier to access
dummies_clean = dummies_clean.to_numpy()
dummies_clean
```

```python
#let write the metadata on the text file
row, col = dummies_clean.shape
with open('problem1.dat.tex', 'w') as f:
    sentence = "DL" + "\n" + "N=" + str(row) +" " + "NM=" + str(col) + "\n" + "FORMAT = FULLMATRIX DIAGONAL PRESENT" +"\n" +"LEVEL LABELS:"   
    f.write(sentence)
    for l in range(0, col):
        f.write('\n')
        f.write('L'+ str(l))
    f.write("\n")
    f.write("Data =")
    f.write("\n")
    
```

```python
#Final Adjacency list and append it to the text file

for k in range(0, col):
    adjacency_matrix_DM_LM = np.zeros((row, row))
    for i in range(0, row):
        for j in range(i,row):
            #print(dummies_clean[i][k],dummies_clean[j][k])
            if dummies_clean[i][k] == dummies_clean[j][k] and i !=j:
                adjacency_matrix_DM_LM[i][j] = 1
                adjacency_matrix_DM_LM[j][i] = 1

    #         else:
    #             print("NO")
    print(adjacency_matrix_DM_LM)
    with open('problem1.dat.tex', 'a') as f:
        data = str(adjacency_matrix_DM_LM)
        data = data.replace("[", "")
        data = data.replace("]", "")
        f.write(data)
        f.write('\n')
```

```python
net=read_ucinet("problem1.dat.tex",couplings="none")
net=transforms.threshold(net,1)
fig=draw(net,
             show=True,
             layout="spring",
             layerColorRule={},
             defaultLayerColor="gray",
             nodeLabelRule={},
             edgeColorRule={"rule":"edgeweight","colormap":"jet","scaleby":0.1}
             )
```

```python
fig.savefig("problem1.pdf")
```

# Problem 2

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python

#rowwise sum
dummies_clean_old['total']=dummies_clean_old.loc[:,:].sum(axis=1)
dummies_value_subject_count = dummies_clean_old.iloc[: , 3:]
dummies_value_subject_count 
```

```python
dummies_value_subject_count = dummies_value_subject_count.to_numpy()
dummies_value_subject_count
```

```python
k = 0
adjacency_matrix_DM_LM = np.zeros((row, row))
for i in range(0, row):
    for j in range(i,row):
         if dummies_value_subject_count[i][k] == dummies_value_subject_count[j][k] and i !=j:
                adjacency_matrix_DM_LM[i][j] = dummies_value_subject_count[i][k]
                adjacency_matrix_DM_LM[j][i] = dummies_value_subject_count[i][k]

#         else:
#             print("NO")
weited_adjacency_matrix = adjacency_matrix_DM_LM.tolist()
weited_adjacency_matrix
```

```python
G = nx.from_numpy_matrix(np.matrix(weited_adjacency_matrix), create_using=nx.DiGraph)
layout = nx.spring_layout(G)
nx.draw(G, layout)
#nx.draw_networkx_edge_labels(G, pos=layout)
plt.show()
```
